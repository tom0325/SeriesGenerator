```python
##獲取 `Series Code` 為 "LOT-NO" 的行
import pandas as pd
df = pd.read_excel("天數測試.xlsx")

test2 = df["Series Code"] == "LOT-NO"

# 找到 `Line No.` 的最大值(int)
max_line_no = df[test2]["Line No."].max()

# 根據 `Series Code` 為 "SOR" 和 `Line No.` 等於最大值過濾 DataFrame
result1 = df[test2 & (df["Line No."] == max_line_no)]
print(result1)
# print(result)

series_code =result1["Series Code"].values[0]
print(series_code)

#建立series code/ line no的list
series_list = []
line_list = []

for i in range(365):
    series_list.append(series_code)
    max_line_no += 10000
    line_list.append(max_line_no)
print(series_list)
print(line_list)

#Check length
if len(series_list)==len(line_list):
    print("Same")
else:
    print("No")
```

```python
##LOT-NO 的Starting Date
import datetime

#Starting date
starting_date = result1["Starting Date"].values[0]
before_starting_date = df.iloc[result1.index-1,2].values[0]

starting_date = pd.to_datetime(starting_date)
before_starting_date = pd.to_datetime(before_starting_date)
print(starting_date)

date_difference = (starting_date - before_starting_date)

print((date_difference))

if date_difference.days == 30 or date_difference.days ==31:
    print("Yes")
elif date_difference.days == 1:
    print("NO")
else:
    print("1")

#Current year
# new_year = starting_date.year + 1
# new_starting_day = starting_date.day + 1
# new_starting_date = pd.Timestamp(year=new_year, month=1, day=new_starting_day)
# print(starting_date)
# print(new_starting_date)

date_list=[]
# 從 2025 年 1 月 1 日開始生成 365 天
for i in range(1, 366):  # 生成一整年的日期
    new_date = starting_date + pd.DateOffset(days=i)
    date_list.append(new_date)

# 將生成的日期轉換為指定格式的字串
formatted_date_list = [date.strftime('%Y/%m/%d') for date in date_list]
print(formatted_date_list)

# 檢查生成的日期長度，應該是 365 天
print(f"生成的日期數量: {len(date_list)}")
```

```python
#Starting No

last_starting_no = result1["Starting No."].values[0]
print(last_starting_no)


#抓取prefix
prefix = str(last_starting_no).split("-")
prefix = prefix[0] + "-"
print("Prefix:"+str(prefix))

#生成新的Starting & Ending
new_starting_no = ""

starting_list = []

for item in range(len(date_list)):
    starting_list.append(new_starting_no)
    starting_list[item]+=prefix
print("Starting list:")
print(starting_list)

for date in range(len(formatted_date_list)):
    # print(formatted_date_list[date])
    year_2 = formatted_date_list[date][2:4] 
    month_2 = formatted_date_list[date][5:7]
    # print(year_2)
    # print(month_2)
    # print(date_2)
    if int(month_2)==10:
        month_2 = "A"
    elif int(month_2)==11:
        month_2 = "B"
    elif int(month_2)==12:
        month_2 = "C"
    date_2 = formatted_date_list[date][8:]
    starting_list[date] += year_2 + month_2 + date_2 + "-0001"
print("Final Starting list:")
print(starting_list)
    


#生成新的Ending
ending_list = []
new_ending_no = ""
for date in range(len(formatted_date_list)):
    # 判斷月份中是否包含 A, B, C（對應10-12月）
    if any(char in starting_list[date] for char in ['A', 'B', 'C']):
        # 拼接前缀到日期部分并保留 "-" 
        new_ending_no = starting_list[date][:len(prefix) + len(year_2) + len(month_2) + len(date_2) + 1] + "9999"
    else:
        # 其他月份正常拼接
        new_ending_no = starting_list[date][:len(prefix) + len(year_2) + len(month_2) + len(date_2)] + "-9999"

    ending_list.append(new_ending_no)

print("Final Ending list:")
print(ending_list)

```

```python
import pandas as pd

# 假設已有的 DataFrame df
df = pd.read_excel("天數測試.xlsx")

# 新的 series_list 資料


# Step 1: 確保 DataFrame 足夠長度來容納新的資料
empty_rows_start = len(df)  # 找到目前df的長度，表示從此行開始填
new_length = empty_rows_start + len(series_list)

# Step 2: 擴展 DataFrame
df = df.reindex(range(new_length))

# Step 3: 插入資料到 'Series Code' 欄位
df.loc[empty_rows_start:, 'Series Code'] = series_list
df.loc[empty_rows_start:, 'Line No.'] = line_list
df.loc[empty_rows_start:, 'Starting Date'] = formatted_date_list
df.loc[empty_rows_start:, 'Starting No.'] = starting_list
df.loc[empty_rows_start:, 'Ending No.'] = ending_list

#存成 df_new
df_new = df
# 檢查結果
print(df_new.head(20))  # 檢查最後幾行是否正確填充
```

```python
# Saving the DataFrame to a new Excel file
output_file = 'New_series_No.xlsx'

# Using ExcelWriter to write DataFrame to an Excel file
with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
    df_new.to_excel(writer, index=False)

print('DataFrame is written to Excel file successfully.')
```
